```python
# Datatype conversion using pandas
df['Date'] = df['Date'].astype(str).apply(lambda x: x.replace('/', '-') if isinstance(x, str) else x)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')

# Encoding categorical features into numerical
df['IsHoliday'] = df['IsHoliday'].map({True: 1, False: 0})
df['Type'] = df['Type'].map({'A': 1, 'B': 2, 'C': 3})

df.tail(3)
```

```python
# make Date feature into --> Day,Month,Year new features and drop Date column

df['Day'] = df['Date'].dt.day
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year
df.drop(columns=['Date'], inplace=True)
df
```

```python
# rearrange the columns order

df= df[['Day','Month','Year','Store','Dept','Type','Weekly_Sales','Size','IsHoliday','Temperature',
           'Fuel_Price', 'MarkDown1', 'MarkDown2','MarkDown3', 'MarkDown4', 'MarkDown5', 'CPI', 'Unemployment']]
df.tail()
```

```python
# create 'markdown' column for --> all markdowns are null means 0, otherwise 1 (any one markdown have value)

df.loc[(df['MarkDown1'].isnull()==True),'MarkDown1'] = 0
df.loc[(df['MarkDown2'].isnull()==True),'MarkDown2'] = 0
df.loc[(df['MarkDown3'].isnull()==True),'MarkDown3'] = 0
df.loc[(df['MarkDown4'].isnull()==True),'MarkDown4'] = 0
df.loc[(df['MarkDown5'].isnull()==True),'MarkDown5'] = 0

df['Markdown'] = df['MarkDown1']+df['MarkDown2']+df['MarkDown3']+df['MarkDown4']+df['MarkDown5']

import numpy as np
df['Markdown'].replace(0, np.nan, inplace=True)
```

```python
x = data.drop(['Weekly_Sales','Day'],axis = 1)
y = data['Weekly_Sales']
x
```

```python
from sklearn import metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25, random_state = 35)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor

model_DecisionTreeRegressor = DecisionTreeRegressor(random_state=42,max_depth=50,min_samples_split=5)#,min_samples_split=10, max_depth=10)4294967295
model_DecisionTreeRegressor.fit(x_train,y_train)
y_pred = model_DecisionTreeRegressor.predict(x_test)


print(f'MAE = {metrics.mean_absolute_error(y_test,y_pred)}')
print(f'MSE = {metrics.mean_squared_error(y_test,y_pred)}')
print(f'RMSE  = {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}')
print(f'R2 Score = {model_DecisionTreeRegressor.score(x_test,y_test)}')
```

```python
from sklearn.ensemble import RandomForestRegressor

model_RandomForest = RandomForestRegressor(n_estimators=300,random_state=10,max_depth=25,min_samples_split=2)
model_RandomForest.fit(x_train, y_train)
y_pred= model_RandomForest.predict(x_test)


print(f'MAE = {metrics.mean_absolute_error(y_test,y_pred)}')
print(f'MSE = {metrics.mean_squared_error(y_test,y_pred)}')
print(f'RMSE  = {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}')
print(f'R2 Score = {model_RandomForest.score(x_test,y_test)}')
```
